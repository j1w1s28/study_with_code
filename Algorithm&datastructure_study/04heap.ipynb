```python
import heapq

scoville =[1, 2, 3, 9, 10, 12]
k = 7
```

```python
heapq.heapify(scoville)
answer = 0
while scoville[0] < k:
    if len(scoville) == 1:
        answer = -1
        break
    else:
        smallest_2 =heapq.nsmallest(2,scoville)
        sum_smallest = smallest_2[0] + smallest_2[1]*2
        heapq.heappop(scoville)
        heapq.heapreplace(scoville,sum_smallest)
        answer += 1
```

```python
def solution(scoville, K):
    import heapq        
    answer = 0
    heapq.heapify(scoville)
    
    if scoville[0] >= K:
        return answer
    
    while scoville[0] < K:
        if len(scoville) == 1:
            return -1
        else:
            #smallest_2 =heapq.nsmallest(2,scoville)
            #sum_smallest = smallest_2[0] + smallest_2[1]*2
            smallest_1=heapq.heappop(scoville)
            smaalest_2 = heapq.heappop(scoville)
            heapq.heappush(scoville,smallest_1 + smaalest_2*2)
            answer += 1
    return answer
```
