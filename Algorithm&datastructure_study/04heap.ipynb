```python
import heapq

scoville =[1, 2, 3, 9, 10, 12]
k = 7
```

```python
heapq.heapify(scoville)
answer = 0
while scoville[0] < k:
    if len(scoville) == 1:
        answer = -1
        break
    else:
        smallest_2 =heapq.nsmallest(2,scoville)
        sum_smallest = smallest_2[0] + smallest_2[1]*2
        heapq.heappop(scoville)
        heapq.heapreplace(scoville,sum_smallest)
        answer += 1
```

```python
def solution(scoville, K):
    import heapq        
    answer = 0
    heapq.heapify(scoville)
    
    if scoville[0] >= K:
        return answer
    
    while scoville[0] < K:
        if len(scoville) == 1:
            return -1
        else:
            #smallest_2 =heapq.nsmallest(2,scoville)
            #sum_smallest = smallest_2[0] + smallest_2[1]*2
            smallest_1=heapq.heappop(scoville)
            smaalest_2 = heapq.heappop(scoville)
            heapq.heappush(scoville,smallest_1 + smaalest_2*2)
            answer += 1
    return answer
```

# 디스크 컨트롤러

```python
jobs =	[[1, 4], [7, 9], [1000, 3]]
```

```python
jobs.sort(key=lambda x : x[0])
```

```python
import copy

jobs_2 = copy.deepcopy(jobs)
jobs_2.sort(key = lambda x : x[1])
```

```python
sum1 = []
sum2 = []
time1 = 0
time2 = 0
```

```python
for i in range(len(jobs)):
    
    if i != 1 and jobs[i-1][1] > jobs[i][0]:
        #앞 프로세스 실행중에 요청이 들어온 경우
        sum1.append(jobs[i][1]+time1-jobs[i][0])
    else:
        #앞 프로세스가 끝나고 요청이 들어온 경우
        sum1.append(jobs[i][1])
        
    if i != 1 and jobs_2[i-1][1] > jobs_2[i][0]:
        sum2.append(jobs_2[i][1]+time2-jobs_2[i][0])
    else:
        sum2.append(jobs_2[i][1])

    time1 += jobs[i][1]
    time2 += jobs_2[i][1]
```

```python
import statistics
avg1= statistics.mean(sum1)
avg2 = statistics.mean(sum2)
```
