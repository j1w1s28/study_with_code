```python
def solution(n, times):
    start,end = 0,max(times)*n
    while start < end:
        mid = (start+end)//2 #mid 값은 모든 사람이 심사 받는데 걸린 시간
        test = 0
        for time in times:
            test += mid // time #주어진 시간에 총 몇명을 심사가능한지 계산
        if test >= n:
            start , end = start, mid
        else:
            start , end = mid+1, end
        
    return start
```
