# 1 같은 숫자는 싫어
배열 arr가 주어집니다. 배열 arr의 각 원소는 숫자 0부터 9까지로 이루어져 있습니다. <br>이때, 배열 arr에서 연속적으로 나타나는 숫자는 하나만 남기고 전부 제거하려고 합니다.<br> 단, 제거된 후 남은 수들을 반환할 때는 배열 arr의 원소들의 순서를 유지해야 합니다.<br> 
### 예를 들면,<br>

- arr = [1, 1, 3, 3, 0, 1, 1] 이면 [1, 3, 0, 1] 을 return 합니다.
- arr = [4, 4, 4, 3, 3] 이면 [4, 3] 을 return 합니다.
배열 arr에서 연속적으로 나타나는 숫자는 제거하고 남은 수들을 return 하는 solution 함수를 완성해 주세요.<br>

### 제한사항
- 배열 arr의 크기 : 1,000,000 이하의 자연수
- 배열 arr의 원소의 크기 : 0보다 크거나 같고 9보다 작거나 같은 정수

```python
def solution(arr):
    answer = []
    
    for i in arr:
        if len(answer) == 0 or answer[-1] != i:
            answer.append(i) 

    return answer
```

```python
from collections import deque
import math
import copy
progresses = [95, 90, 99, 99, 80, 99]
speeds = 	[1, 1, 1, 1, 1, 1]
```

```python
left =deque([math.ceil((100-progresses[i])/speeds[i])  for i in range(len(progresses))])
```

```python
def func(arrs,start_index=0,answer= []):
    count = 1    
    print(start_index)
    for i in range(start_index,len(arrs)):        
        for j in range(i+1,len(arrs)):
            if arrs[i] >= arrs[j]:
                count += 1
            else :
                answer.append(count)
                #print(j)
                return func(arrs=arrs,start_index=j,answer=answer)        
    answer.append(count)
    return answer
```
