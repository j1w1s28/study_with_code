# STACK Design

```python
class Stack():
    def __init__(self):
        self.data = []
        self.top = -1 #현재 가장 마지막으로 들어와있는 값의 index
    
    def push(self,x):
        #insert X
        self.data.append(x)
        self.top += 1

    def peek(self):
        #get item
        if not self.is_empty():
            return self.data[self.top]
        else: return None
            

    def pop(self):
        #delete an item
        if not self.is_empty():
            del self.data[self.top]
            self.top -= 1

    def is_empty(self):
        return (self.top == -1)
```
