```python
class HashTable:
    def __init__(self):
        self.capacity = 1001
        self.buckets = [None] * self.capacity
        self.size = 0

    def hash(self,key):
        return key % self.capacity

    def insert(self,key,value):
        index = self.hash(key)
        node = self.buckets[index]

        new_node = Node(key,value)
        new_node.next = node
        self.buckets[index] = new_node
        self.size += 1

        #load_factor 계산 후 rehash
        load_factor = self.size / self.capacity
        if load_factor >= 0.8:
            self.rehash()
    
    def retrieve(self,key):
        index =self.hash(key)
        node = self.buckets[index]

        while node is not None and node.key != key:
            node = node.next

        if node is None: # Not exist
            return None
        else: #Found
            return node.value
    def delete(self,key):
        index = self.hash(key)
        node = self.buckets[index]

        prev = None
        while node is not None and node.key != key:
            prev = node
            node = node.next

        if node is None: # Not exist
            return None
        else: #Found
            result = node.value

            if prev is None:
                node =None
            else:
                prev.next = prev.next.next
            self.size -= 1
            return result
    def rehash(self):
        self.capacity *= 2
        new_buckets = [None] * self.capacity

        for node in self.buckets:
            while node is not None:
                index = self.hash(node.key)
                new_node = Node(node.key,node.value)
                new_node.next = new_buckets[index]
                new_buckets[index] = new_node
                node = node.next

        self.buckets = new_buckets
        return
    
class Node:
    def __init__(self,key,value):
        self.key = key
        self.value = value
        self.next = None
        
        
```
