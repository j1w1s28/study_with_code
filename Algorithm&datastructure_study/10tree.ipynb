# 트리 순회

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
```

```python
def preorder(node):
    if node:
        print(node.value, end="")
        preorder(node.left)
        preorder(node.right)

def inorder(node):
    if node:
        inorder(node.left)
        print(node.value, end="")
        inorder(node.right)

def postorder(node):
    if node:
        postorder(node.left)
        postorder(node.right)
        print(node.value, end="")
```

```python
def find_node(root, target_value):
    if root is None:
        return None
    
    if root.value == target_value:
        return root
    
    left_result = find_node(root.left, target_value)
    if left_result:
        return left_result
    
    right_result = find_node(root.right, target_value)
    return right_result
```

```python
root = TreeNode('A')
n = int(input())
for i in range(n):
    node, left, right = map(str,input().split())
    cur_node = find_node(root,node)
    cur_node.left = TreeNode(left) if left != '.' else None
    cur_node.right = TreeNode(right) if right != '.' else None
```

```python
print("Preorder Traversal:")
preorder(root)
print("\n")
print("Inorder  Traversal:")
inorder(root)
print("\n")
print("Postorder  Traversal:")
postorder(root)
print("\n")
```

```python
def build_tree_from_inorder_postorder(inorder, postorder):
    if not inorder or not postorder:
        return None
    
    root_value = postorder.pop()
    root = TreeNode(root_value)
    
    inorder_index = inorder.index(root_value)
    
    root.right = build_tree_from_inorder_postorder(inorder[inorder_index+1:], postorder)
    root.left = build_tree_from_inorder_postorder(inorder[:inorder_index], postorder)
    
    return root
```

```python
n = 3
inorder = ['1','2','3']
postorder = ['1','3','2']
def build_tree_from_inorder_postorder(inorder, postorder, start, end):
    if start > end:
        return None
    
    root_value = postorder.pop()
    root = TreeNode(root_value)
    
    inorder_index = inorder.index(root_value, start, end + 1)
    
    root.right = build_tree_from_inorder_postorder(inorder, postorder, inorder_index + 1, end)
    root.left = build_tree_from_inorder_postorder(inorder, postorder, start, inorder_index - 1)
    
    return root
```

```python
class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, value):
        self.root = self._insert_recursive(self.root, value)

    def _insert_recursive(self, root, value):
        if root is None:
            return TreeNode(value)
        
        if value < root.value:
            root.left = self._insert_recursive(root.left, value)
        elif value > root.value:
            root.right = self._insert_recursive(root.right, value)
        
        return root

    def inorder_traversal(self, root):
        if root:
            self.inorder_traversal(root.left)
            print(root.value, end=" ")
            self.inorder_traversal(root.right)

    def postorder_traversal(self, root):
        if root:
            self.postorder_traversal(root.left)
            self.postorder_traversal(root.right)
            print(root.value, end=" ")
# Given preorder numbers
preorder_numbers = [10, 5, 3, 7, 15, 12, 18]

# Create a Binary Search Tree from the given preorder numbers
bst = BinarySearchTree()
for num in preorder_numbers:
    bst.insert(num)

print("Given preorder numbers:", preorder_numbers)

# Perform inorder traversal to see the sorted values in the BST
print("Inorder Traversal (Sorted):")
bst.inorder_traversal(bst.root)

print('Postorder Traversal')
bst.postorder_traversal(bst.root)
```

```python
def construct_postorder(preorder):
    if not preorder:
        return []
    
    root_value = preorder[0]
    left_subtree = []
    right_subtree = []
    
    for val in preorder[1:]:
        if val < root_value:
            left_subtree.append(val)
        else:
            right_subtree.append(val)
    
    postorder_left = construct_postorder(left_subtree)
    postorder_right = construct_postorder(right_subtree)
    
    return postorder_left + postorder_right + [root_value]
```
